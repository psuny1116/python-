# str.startswith(문자열, 시작위치, 끝위치)

- 괄호 안에 적은 문자열로 시작하는지 확인

```python
string = 'hello python hi how are you'

print(string.startswith('hello'))
print('찾는 범위: ', string[16 : 19])
print(string.startswith('how', 16, 19))
print('찾는 범위: ', string[2 : 10])
print(string.startswith('how', 2, 10))
```
