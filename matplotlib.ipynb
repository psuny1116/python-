```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl
import numpy as np
import pandas as pd

mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
from matplotlib import rcParams

import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
# 데이터 생성
data = np.array([1, 3, 5, 7, 9, 11, 13, 15])

plt.plot(data)
plt.show()
```

```python
# x축, y축 값 지정

x = np.array([10, 20, 30, 40, 50, 60, 70, 80])
y = np.array([1, 3, 5, 7, 9, 11, 13, 15])
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, color = 'r')
plt.show()
```

```python
plt.plot(x, y, color = 'r', linestyle = '--')
plt.show()
```

```python
plt.plot(x, y, 'rs--')
plt.show()
```

```python
plt.plot(x, y,
        color = 'darkblue',
        linewidth = 3,
        linestyle = '--',
        marker = 'o',
        ms = 15,
        mec = 'orange',
        mew = 3,
        mfc = 'yellow')
plt.show()
```

```python
# 그래프 전체 스타일 목록

print(plt.style.available)
```

```python
# dark_background 스타일

with plt.style.context('grayscale'):
    plt.title('grayscale 스타일')
    plt.plot(x, y, color = 'darkblue', linewidth = 3, linestyle = '--', marker = 'o', ms = 15, mec = 'orange', mew = 3, mfc = 'yellow')
plt.show()
```

```python
# fivethirtyeight 스타일

with plt.style.context('fivethirtyeight'):
    plt.title('fivethirtyeight 스타일')
    plt.plot(x, y, color = 'darkblue', linewidth = 3, linestyle = '--', marker = 'o', ms = 15, mec = 'orange', mew = 3, mfc = 'yellow')
plt.show()
```

```python
# figure 크기 지정

plt.figure(figsize = (15, 5))     # 가로길이, 세로길이
plt.plot(x, y, color = 'darkblue', linewidth = 3, linestyle = '--', marker = 'o', ms = 15, mec = 'orange', mew = 3, mfc = 'yellow')
plt.show()
```

```python
# plot 범위 설정
## plt.xlim(최소, 최대), plt.ylim(최소, 최대)

plt.plot(x, y, color = 'darkblue', linewidth = 3, linestyle = '--', marker = 'o', ms = 15, mec = 'orange', mew = 3, mfc = 'yellow')
plt.xlim(0, 50)
plt.ylim(0, 10)
plt.show()
```

```python
# 축, 제목 이름

plt.figure(figsize = (15, 5))
plt.rc('font', size = 20)     # 폰트 사이즈
plt.plot(x, y, color = 'darkblue', linewidth = 3, linestyle = '--', marker = 'o', ms = 15, mec = 'orange', mew = 3, mfc = 'yellow')
plt.xlabel('x축')
plt.ylabel('y축')
plt.title('그래프')
plt.show()
```

```python
# tick 설정(축에 표시되는 지점. 보여지고 싶은 간격 설정할때 사용)

plt.plot(x, y, color = 'darkblue', linewidth = 3, linestyle = '--', marker = 'o', ms = 15, mec = 'orange', mew = 3, mfc = 'yellow')
plt.xticks(np.linspace(0, 80, 5))
plt.yticks(np.linspace(0, 15, 3))
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 200)
y1, y2 = np.cos(x), np.sin(x)

plt.plot(x, y1)
plt.plot(x, y2)     # plt.plot을 여러번 하면 하나의 figure에 여러개의 그래프가 나타남
plt.show()
```

```python
# legend 설정
## plt.plot(label = '범례에 표시될 이름') / plt.legend(loc = '범례 위치')

x = np.linspace(-np.pi, np.pi, 200)
y1, y2 = np.cos(x), np.sin(x)

plt.plot(x, y1, label = '코사인 그래프')
plt.plot(x, y2, label = '사인 그래프')
plt.legend(loc = 'upper left')
plt.show()
```

## 다중 그래프(subplot)

- plt.subplot('행 axes 개수', '열 axes 개수', '그래프 시각화 순서')

```python
# 그래프 2개

x1 = np.array([10, 20, 30, 40, 50, 60, 70, 80])
x2 = np.array([10, 20, 30, 40, 50, 60, 70, 80])
y1 = np.array([1, 3, 5, 7, 9, 11, 13, 15])
y2 = np.flip(y1)

plt.figure(figsize = (10, 10))     # figure 생성

ax1 = plt.subplot(2, 1, 1)     # ax1은 2x1에서 첫번째 그래프
plt.plot(x1, y1, 'yo-')
plt.title('첫번째 그래프')

ax2 = plt.subplot(2, 1, 2)     # ax2은 2x1에서 두번째 그래프
plt.plot(x2, y2, 'r.-')
plt.title('두번째 그래프')

plt.show()
```
