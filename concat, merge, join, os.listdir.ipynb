# concat

```python
import pandas as pd

df1 = pd.DataFrame({'A': [1, 3, 5, 7, 9],
                   'B': [11, 12, 13, 14, 15],
                   'C': [0, 1, 2, 3, 4],
                   'D': [101, 102, 103, 104, 105]}, index = [1, 2, 3, 4, 5])
df2 = pd.DataFrame({'A': [2, 4, 6, 8, 10],
                   'B': [1, 2, 3, 4, 5],
                   'C': [30, 31, 32, 33, 34],
                   'D': [1011, 1021, 1031, 1041, 1051]}, index = [6, 7, 8, 9, 10])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis = 1)     # 두 데이터의 index가 일치하지 않으므로 아래와 같은 모양으로 결합됨
```

```python
df1.reset_index(drop = True, inplace = True)
df2.reset_index(drop = True, inplace = True)
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2], axis = 1)
```

## 폴더에 있는 여러 파일 한번에 병합

```python
import os
os.listdir('일별 오염 데이터')
```

```python
# 1. for문 사용: 코드가 김, 비효율적, 하지만 메모리 문제가 생길 가능성 적음

merged_df = pd.DataFrame()

for file in os.listdir('일별 오염 데이터'):
    if '오염_수준.txt' in file:
        df = pd.read_csv('일별 오염 데이터/' + file, sep = '\t', engine = 'python')
        merged_df = pd.concat([merged_df, df], axis = 0, ignore_index = True)
        
merged_df.head()
```

```python
merged_df.shape
```

```python
# 2. list comprehension 사용: 코드 짧음. 효율적, 하지만 메모리 문제가 생길 수 있음(위와 결과는 동일)

merged_df2 = pd.concat([pd.read_csv('일별 오염 데이터/' + file, sep = '\t', engine = 'python') 
                       for file in os.listdir('일별 오염 데이터') if '오염_수준.txt' in file])
merged_df2.head()
```

```python
merged_df2.shape
```

# merge

```python
df1 = pd.DataFrame({'id': ['X01', 'X02', 'X03', 'X04', 'X05'],
                   'A': [1, 3, 5, 7, 9],
                   'B': [11, 12, 13, 14, 15],
                   'C': [0, 1, 2, 3, 4],
                   'D': [101, 102, 103, 104, 105]})
df2 = pd.DataFrame({'id': ['X01', 'X02', 'X03', 'X04', 'X06'],
                   'E': [2, 4, 6, 8, 10],
                   'F': [1, 2, 3, 4, 5],
                   'G': [30, 31, 32, 33, 34],
                   'H': [1011, 1021, 1031, 1041, 1051]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on = 'id')
```

```python
pd.merge(df1, df2, on = 'id', how = 'inner')
```

```python
pd.merge(df1, df2, on = 'id', how = 'outer')
```

```python
pd.merge(df1, df2, on = 'id', how = 'left')
```

```python
pd.merge(df1, df2, on = 'id', how = 'right')
```

```python
df3 = pd.DataFrame({'user_id': ['X01', 'X02', 'X03', 'X04', 'X06'],
                   'E': [2, 4, 6, 8, 10],
                   'F': [1, 2, 3, 4, 5],
                   'G': [30, 31, 32, 33, 34],
                   'H': [1011, 1021, 1031, 1041, 1051]})
```

```python
# key 칼럼명이 다를 경우

pd.merge(df1, df3, left_on = 'id', right_on = 'user_id')     # left_on, right_on 키가 모두 표에 나타나므로 하나는 삭제해야 함
```

```python
df1 = pd.read_csv('merge_실습_employee_group.csv', engine = 'python')
df1.head()
```

```python
df2 = pd.read_csv('merge_실습_name_hire_date.csv', engine = 'python', index_col = 'name')
df2.head()
```

```python
pd.merge(df1, df2, left_on = 'employee', right_index = True)     # right_index = True: 오른쪽 데이터프레임의 인덱스를 key변수로 사용
```

# join

```python
df1 = pd.DataFrame({'A': [1, 3, 5, 7, 9],
                   'B': [11, 12, 13, 14, 15],
                   'C': [0, 1, 2, 3, 4],
                   'D': [101, 102, 103, 104, 105]}, index = [1, 2, 3, 4, 5])
df2 = pd.DataFrame({'E': [2, 4, 6, 8, 10],
                   'F': [1, 2, 3, 4, 5],
                   'G': [30, 31, 32, 33, 34],
                   'H': [1011, 1021, 1031, 1041, 1051]}, index = [6, 7, 8, 9, 10])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)     # index가 달라서 원하는 대로 결합이 안됨
```

```python
df1.reset_index(drop = True, inplace = True)
df2.reset_index(drop = True, inplace = True)
```
