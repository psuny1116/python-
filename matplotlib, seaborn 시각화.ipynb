# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 10)     # 0 ~ 5를 10등분
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x, y, 'g')     # g: 초록색
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.7, 1])     # 위치: 왼쪽, 아래쪽, 너비, 높이

axes.plot(x, y, 'r')
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.title('plot title')
```

```python
# 두 플랏을 겹쳐 그리기

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes2 = fig.add_axes([0.2, 0.6, 0.3, 0.3])

axes1.plot(x, y, 'r')
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.title('plot title1')

axes2.plot(y, x, 'b')
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.title('plot title2')
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, 'g')
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.title('plot title')
```

```python
# 2개를 그릴 때 subplot 사용

fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y, 'r')
axes[1].plot(y, x, 'b')

axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('title1')

axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('title2')
```

```python
fig.savefig('graph.png')
```

# Seaborn - Distribution Plot

```python
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
data.head()
```

## Distribution Plot

```python
sns.distplot(data['total_bill'])
```

```python
sns.distplot(data['total_bill'], kde = False)
```

```python
sns.distplot(data['total_bill'], kde = False, bins = 10)
```

## Joint Plot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = data)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = data, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = data, kind = 'reg')
```

## Pair Plot

```python
sns.pairplot(data)
```

```python
sns.pairplot(data, hue = 'smoker')
```

## Rug Plot

```python
sns.rugplot(data['total_bill'])
```

## Facet Grid

```python
data = sns.load_dataset('tips')
data.head()
```

```python
data['time'].unique()
```

```python
g = sns.FacetGrid(data, col = 'time', row = 'smoker')     # 카테고리 칼럼만 적용 가능
g = g.map(plt.hist, 'total_bill')
```

```python
g = sns.FacetGrid(data, col = 'time', row = 'smoker')
g = g.map(plt.scatter, 'total_bill', 'tip')
```

```python
g = sns.FacetGrid(data, col = 'time', row = 'smoker', hue = 'sex')
g = g.map(plt.scatter, 'total_bill', 'tip').add_legend()
```
