# pd.melt()

- pd.melt(dataframe, id_vars, value_vars, var_name, value_name)
- 여러 개의 열을 하나의 열 안에 다수의 행으로 넣음
- id_vars: 위치를 유지할 열 이름(범주형)
- value_vars: 가공할 열 이름(범주형). id_vars로 지정한 열 외에 나머지 모든 열을 넣을 것이라면 따로 지정할 필요 없음
- var_name: 가공되어 새로 만들어지는 열 이름을 지정(디폴트 = variable)
- value_name: 가공할 열에 해당하는 수치형 데이터값의 열 이름 지정(디폴트 = value)

```python
import pandas as pd
import seaborn as sns

data = pd.read_csv('pew.csv')
data.head()
```

- religion: 종교
- 그 외 열: 소득수준
- 데이터: 각 종교 및 소득수준에 해당하는 사람들의 수

```python
# religion열은 그대로 두고 소득수준을 income이라는 열로 데이터프레임 변환

pd.melt(data, id_vars = 'religion', var_name = 'income', value_name = 'cnt')
```
