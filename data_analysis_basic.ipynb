```python
# 수학 칼럼이 80보다 작은 경우만 선택

cond = data['수학'] < 80
data[cond]
```

```python
# True/ Fasle 값을 가지고 조건을 만들어도 됨. 이때, 데이터 프레임의 행 개수와 리스트의 원소 개수가 동일해야 함
cond = [True, False, True]     # 이렇게 하면 첫번째, 세번째 행만 가져옴
data[cond]
```

```python
# 영어나 수학 둘 중 하나라도 80을 넘는 경우

cond3 = (data['영어'] > 80)
cond4 = (data['수학'] > 80)

data[cond3 | cond4]     # | = or, & = and
```

```python
# 영어, 수학 모두 80넘는 경우

data[cond3 & cond4]
```

```python
# \(역슬러시)를 이용하면 코드 중간에 줄바꿈을 할 수 있음

cond = (data['영어'] >= 70) \
     & (data['수학'] >= 70) \
     & (data['수학'] < 90)

data[cond]
```

```python
# 첫번째 데이터

fpath = './데이터분석/인강/6. 러닝스푼즈_포트폴리오/파이썬 기초 강의자료/Ch2)Pandas_Basic/data/exam.xlsx'
A = pd.read_excel(fpath, index_col = '번호')     # index_col로 인덱스로 사용할 컬럼 지정
A
```

```python
# 두번째 데이터

fpath2 = './데이터분석/인강/6. 러닝스푼즈_포트폴리오/파이썬 기초 강의자료/Ch2)Pandas_Basic/data/exam_extra.xlsx'
B = pd.read_excel(fpath2, index_col = '번호')     # index_col로 인덱스로 사용할 컬럼 지정
B
```

#### 데이터 병합: pd.merge(A, B, how = 'left', left_on = 'A칼럼명', right_index = True)
* how:
    - left(왼쪽표A 기준)
    - right(오른쪽표B 기준)
    - inner(A, B 둘 다 있는 데이터만)
    - outter(A, B 한쪽이라도 있는 데이터)

* left_on: 왼쪽 테이블 A 병합 기준 칼럼 지정 / left_index = True: 왼쪽 테이블 A 병합 기준 인덱스 지정 (둘 중 1개만 써야 함)
* right_on: 오른쪽 테이블 B 병합 기준 칼럼 지정 / right_index = True: 오른쪽 테이블 B 병합 기준 인덱스 지정 (둘 중 1개만 써야 함)
* on: A & B 같은 이름의 열을 기준으로 지정할 경우

```python
# A 테이블 기준으로 병합

total = pd.merge(A, B, how = 'left', left_index = True, right_index = True)
total
```

```python
pd.merge(A, B, how = 'right', left_index = True, right_index = True)     # B 테이블 기준으로 병합
```

```python
pd.merge(A, B, how = 'inner', left_index = True, right_index = True)     # A, B 테이블 양쪽 모두 존재
```

```python
pd.merge(A, B, how = 'outer', left_index = True, right_index = True)     # A, B 양쪽에 하나라도 존재
```

```python
# 현재 위치에 데이터 저장

total.to_excel('./total.xlsx', index = False)     # index = False를 쓰면 저장 시에 인덱스 저장 안함. 생략 가능
```
