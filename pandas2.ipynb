# join

- 두 개의 데이터 프레임을 인덱스 기준으로 병합
- df1.join(df2, on = '키 인덱스', how = '조인 방법', sort = '정렬 여부')


- join을 사용하기 위해서는 키로 지정할 열을 인덱스로 변경해야 함
- set_index('키로 지정할 열')

```python
df3 = df2.set_index('Korean Name')
df3
```

```python
df1.join(df3, on = 'Korean Name', how = 'left')
```

```python
df1.join(df3, on = 'Korean Name', how = 'right')
```

```python
df1.join(df3, on = 'Korean Name', how = 'inner')
```

```python
df1.join(df3, on = 'Korean Name', how = 'outer')
```

# merge

- 두 개의 데이터프레임을 겹치는 값을 기준으로 병합
- pd.merge(df1, df2, on = '키 컬럼', how = '조인 방법', sort = '정렬 여부')

```python
pd.merge(df1, df2, on = 'Korean Name', how = 'left')
```

```python
pd.merge(df1, df2, on = 'Korean Name', how = 'right')
```

```python
pd.merge(df1, df2, on = 'Korean Name', how = 'inner')
```

```python
pd.merge(df1, df2, on = 'Korean Name', how = 'outer')
```

# concat

- 두 개의 데이터프레임을 행이나 열 방향으로 단순히 병합
- 행방향 병합
    - pd.concat([df1, df2], axis = 0)
- 열방향 병합
    - pd.concat([df1, df2], axis = 1)

```python
# 행 방향 병합

pd.concat([df1, df2], axis = 0)
```

```python
# 열 방향 병합

pd.concat([df1, df2], axis = 1)
```

```python
# 중복되는 값만 병합(join 활용)

pd.concat([df1, df2], axis = 1, join = 'inner')
```

# 자료형 변환

```python
df = df.iloc[:, 1:11].head(7)
df = df[['Korean Name', 'Date of Birth', 'Height', 'Weight']]
df
```

```python
df.info()
```

```python
# Height 컬럼의 타입을 int에서 str로 변환

df['Height'] = df['Height'].astype('str')
```

```python
df.info()
```

```python
# Height 컬럼의 타입을 str에서 int로 변환

df['Height'] = df['Height'].astype('int')
```

```python
df.info()
```

```python
# Date of Birth 컬럼의 타입을 object에서 datetime으로 변환

df['Date of Birth'] = pd.to_datetime(df['Date of Birth'])
```

```python
df.info()
```

```python
df['Year of Birth'] = df['Date of Birth'].dt.year     # 연도 추출
df['Month of Birth'] = df['Date of Birth'].dt.month     # 월 추출
df['Day of Birth'] = df['Date of Birth'].dt.day     # 일 추출
df['Day of week of Birth'] = df['Date of Birth'].dt.dayofweek     # 요일 추출(월요일: 0 ~ 일요일: 6)
df
```

#### datetime 연산

```python
df['Day of Birth'] = pd.to_timedelta(df['Day of Birth'], unit = 'day')
df
```

```python
df['Date of Birth'] + df['Day of Birth']
```

# 결측값

```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://bit.ly/3gRXTfD')
df = df.iloc[15:30]
df = df[['K. Stage Name', 'Date of Birth', 'Group', 'Height', 'Weight', 'Birthplace']]
df
```

```python
# 0값을 Nan으로 변환

df = df.replace(0, np.NaN)
df
```

```python
df.info()
```

```python
# 결측치 확인

df.isnull()
```

```python
df.notnull()
```

```python
# 결측치 갯수 확인

df.isnull().sum()
```

```python
# 결측치 제거(dropna)

df.dropna()
```

```python
# subset = ['컬럼명']을 지정하면 해당 컬럼의 결측치 존재여부에 따라 달라짐

df.dropna(subset = ['Group'])
```

```python
df.dropna(subset = ['Group', 'Height'])
```

```python
# 특정값으로 결측치 대체(fillna)

df['Height'] = df['Height'].fillna(160)
df
```

```python
# 평균값으로 대체

df['Weight'] = df['Weight'].fillna(df['Weight'].mean())
df
```

```python
# 최빈값으로 대체

df['Birthplace'].value_counts()
```

```python
df['Birthplace'].value_counts().index
```

```python
df['Birthplace'].value_counts().index[0]
```

```python
df['Birthplace'] = df['Birthplace'].fillna(df['Birthplace'].value_counts().index[0])
df
```

```python
# 인덱스 재정렬(reset_index)

df = df.reset_index(drop = True)
df
```

# 중복값

```python
# 중복 데이터 임의로 생성

df = df.append(df.iloc[-1]).reset_index(drop = True)
df
```

```python
# 중복 데이터 탐색(duplicated())
## 중복된 행 찾기

df[df.duplicated()]
```

```python
# 특정 열에 중복값 찾기

df['Group'].duplicated()
```

```python
# 중복 데이터 제거(drop_duplicates())

df.drop_duplicates()     # 완벽히 중복된 행 제거
```

```python
# 특정 열을 기준으로 중복값 제거

df.drop_duplicates(subset = ['Group'])
```

```python
# 특정 열을 기준으로 중복값 제거(마지막 값 남기기)

df.drop_duplicates(subset = ['Group'], keep = 'last')
```

# apply

```python
df = pd.read_csv('https://bit.ly/3gRXTfD')
```

```python
df = df.loc[df['Gender'] == 'M'][:3].append(df.head(9)).reset_index(drop = True)
df = df[['Height', 'Weight', 'Gender']]
df
```

```python
def replace_zero(x):
    if x == 0:
        x = 165
    return x
```

```python
df['Height'] = df['Height'].apply(replace_zero)
df
```

# lambda

- lambda '입력변수': '리턴값' '조건문(옵션)'

```python
# 값이 0일 경우 원하는 값으로 대체

df['Weight'] = df['Weight'].apply(lambda x: 50 if x == 0 else x)
df
```

```python
# 파생 컬럼 생성

df['Weight/Height'] = df.apply(lambda x: x['Weight']/x['Height'], axis = 1)
df
```

# map

- 데이터 값을 특정한 값으로 매핑하고 싶은 경우 사용
- 반복문을 이용하여 값을 치환하는 방법보다 효율적
- 하나의 컬럼인 series형으로만 적용 가능
- df['컬럼명'].map('매핑정보')

```python
map_info = {'M': '남자', 'F': '여자'}

df['Gender'] = df['Gender'].map(map_info)
df
```
