# 데이터 분석 과정

## 1. 데이터 수집

- 데이터베이스
    - 여러 사람이 공유하여 사용할 목적으로 체계화 해, 통합 관리하는 데이터의 집합
    - 예: 관계형 DB(오라클, my sql), NoSQL(몽고DB)

- 서버 로그 분석
    - 데이터베이스에 쌓이는 데이터와는 별개로 컴퓨터 네트워크가 존재하는 서버상에 쌓이는 로그들
    - 예: ip 접속기록, 페이지뷰 기록 등
    
- 크롤링

## 2. EDA

- 탐색적 데이터 분석
    - 데이터의 직관을 얻기 위하여 그래프 혹은 통계적 지식을 통해 살펴보는 방법
    - 데이터 확인 및 분석 설계 단계
    - EDA에는 시각화가 많이 사용되지만, 데이터 시각화와는 다름
    - EDA에서의 시각화는 데이터가 어떤 목적, 형태로 존재하는지 확인을 위함일 뿐

## 3. 전처리

- 데이터 가공
    - 분석을 위한 데이터 가공
    - 전체 과정 중 약 50% 차지

- 준비 단계
    - EDA, 시각화, 피처엔지니어링을 위한 준비 단계

- 이상치 및 결측치 전처리
    - 이상치: 정상 범주에서 크게 벗어난 값으로써 이상치를 포함하여 분석을 진행할 경우 분석 결과가 왜곡될 가능성 존재
    - 결측치: 누락값, 비어 있는 값으로 데이터 수집 당시에 기록되지 않거나 누락된 값을 의미
    - 제거: 이상치 및 결측치가 포함되어 있는 행 혹은 열 제거
    - 대체: 이상치 및 결측치를 해당 칼럼의 최대값, 평균값, 중앙값 등으로 대체
    - 예측: 이상치 및 결측치가 포함된 칼럼의 특성을 고려하여 예측값으로 채워 넣음
    
- 라벨 인코딩
    - 머신러닝을 학습시키기 위해선 숫자가 아닌 단어 혹은 문자열을 숫자로 바꿔줘야만 학습이 가능
    - 라벨 인코딩은 문자열을 카테고리별로 적용시키는 인코딩 방식
    
- 원핫 인코딩
    - 문자열을 숫자로 바꾼다는 점은 라벨 인코딩과 같지만
    - 원핫 인코딩은 문자열을 카테고리별로 바꾸는 것이 아닌 0과 1만 소유할 수 있는 테이블을 만들어 해당되는 데이터만 1로 표시하고 남은 데이터는 0으로 채우는 인코딩 방식

# 프로젝트

- 하나의 주제를 통한 일련의 분석 과정
- 프로젝트 내에는 분석과정이 모두 담겨 있어야 함
- 분석 과정은 모두 담아내고 하나의 프로젝트의 결과를 보고할 때에는 핵심 부분만 요약해서 정리

## 프로젝트 요약

- 데이터 소개 및 분석 동기
    - 데이터 구성/ 칼럼의 의미
- 분석을 하며 생긴 한계점
- 한계점 해결 과정 또는 개선 방향
- EDA의 산출물(인사이트 구축 과정, 시각화, 가설 및 검정)
- 기술 스택(사용 언어, 사용 기법 등)|

# 포트폴리오
