# pivot_table

```python
df = pd.read_csv('온라인_판매기록.csv', encoding = 'cp949')
df.head()
```

```python
pt1 = pd.pivot_table(df,     # 데이터
                    index = '제품',     # 행에 들어갈 조건
                    columns = '쇼핑몰 유형',     # 열에 들어갈 조건
                    values = '판매금액',     # 집계 대상 컬럼 목록
                    aggfunc = 'mean')     # 집계 함수
pt1
```

```python
pt2 = pd.pivot_table(df,
                    index = ['제품'],
                    columns = ['쇼핑몰 유형'],
                    values = ['판매금액', '수량'],
                    aggfunc = 'max')
pt2
```

# groupby

- pivot_table과 groupby의 공통점: 둘 다 조건부 통계량을 기준으로 데이터를 집계함
- pivot_table과 groupby의 차이점: 출력 구조에 차이가 있음
- 예) pivot_table

. |남성|여성
-|-|-
대졸|200|300
고졸|300|150

- 예) groupby

학력|성별|소득
-|-|-
대졸|남성|200
 |여성|300
고졸|남성|300
|여성|150

- 보통 출력 결과 자체가 결과물인 경우에는 pivot_table을
-                       중간 산출물인 경우에는 groupby를 사용

```python
df = pd.read_csv('온라인_판매기록.csv', encoding = 'cp949')
df.head()
```

```python
df.groupby(['쇼핑몰 유형'])['수량'].mean()     # 분할기준 컬럼, 적용 기준 컬럼, 집계함수
```

```python
df.groupby(['쇼핑몰 유형'], as_index = False)['수량'].mean()     # as_index: 분할 기준 컬럼을 인덱스로 사용할 것인지 여부
```

```python
df.groupby(['쇼핑몰 유형'])['수량', '판매금액'].agg(['mean', 'max']) # 여러 집계 함수나 사용자 정의 함수를 쓰기 위해선 agg함수 사용
```

```python
def my_func(value):
    return max(value) - min(value)

df.groupby(['쇼핑몰 유형', '제품'])['수량', '판매금액'].agg(my_func) # 여러 집계 함수나 사용자 정의 함수를 쓰기 위해선 agg함수 사용
```
