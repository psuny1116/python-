```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl
import numpy as np

mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
from matplotlib import rcParams
```

## bar 그래프

- x, y 중 하나의 값이 카테고리형일 경우 bar그래프
- plt.bar(x, y, color = ['색상'], width = '폭')
- 가로 방향으로 그리려면 barh

```python
x = np.arange(5)
y = [80, 70, 40, 90, 60]
xlabel = ['국어', '수학', '영어', '과학', '미술']
```

```python
# 세로막대

plt.bar(x, y)
plt.xticks(x, xlabel)
plt.ylabel('점수')
plt.show()
```

```python
# 가로막대

plt.barh(x, y)
plt.yticks(x, xlabel)
plt.xlabel('점수')
plt.show()
```

```python
# 여러개의 bar 그래프(나란한 형태)
## bar의 폭 크기를 지정한 후, 수동으로 다음 그려질 그래프 위치를 지정

x = np.arange(5)
women = [80, 70, 40, 90, 60]
men = [70, 80, 45, 75, 80]
xlabel = ['국어', '수학', '영어', '과학', '미술']

bar_width = 0.35     # 폭 설정
plt.bar(x, women, width = bar_width, label = '여자')
plt.bar(x + bar_width, men, width = bar_width, label = '남자')     # 두번째 그래프: 인덱스+폭
plt.xticks(x + bar_width / 2, xlabel)     # x축에 보여질 문구 위치
plt.ylabel('점수')
plt.legend(loc = 'upper right')
plt.show()
```

```python
# 여러개의 bar 그래프(stacked 형태)
## bottom 옵션 사용

bar_width = 0.35
plt.bar(x, women, width = bar_width, label = '여자', bottom = men)     # 아래에 men이 위치하도록 설정
plt.bar(x, men, width = bar_width, label = '남자')
plt.xticks(x, xlabel)
plt.ylabel('점수')
plt.legend(loc = 'upper right')
plt.show()
```

## pie 그래프

- 카테고리별 값의 상대적인 비교
- plt.pie(카테고리값, color = ['색상'], width = '폭', explode = '돌출 옵션')

```python
labels = ['애플', '삼성', '오포', '샤오미', '기타']
sizes = [16, 21, 16, 14, 33]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'lightgray']

plt.figure(figsize = (5, 5))
plt.pie(sizes, labels = labels)
plt.title('휴대폰 시장 점유율')
plt.show()
```

```python
# 파이 돌출 옵션

explode = (0, 0.1, 0, 0, 0)

plt.figure(figsize = (5, 5))
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%0.1f%%', shadow = True)
plt.title('휴대폰 시장 점유율')
plt.show()
```

## histogram

- 빈도 분포 확인
- plt.hist('발생 빈도가 있는 값')

```python
x = np.random.randn(2000)
plt.hist(x)
plt.show()
```

```python
# bin 크기 지정

plt.hist(x, bins = 100)
plt.show()
```

```python
# 누적 분포

plt.hist(x, bins = 100, cumulative = True)
plt.show()
```

## scatter

- 두 데이터 집합의 상관관계 확인
- plt.scatter(x, y)

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.show()
```

```python
# 옵션

colors = np.random.rand(50)
area = np.random.rand(50) * 500

plt.scatter(x, y, sizes = area, c = colors, alpha = 0.5)
plt.show()
```

## imshow(이미지 그래프)

- 픽셀의 모임이 행과 열을 가진 2차원 데이터
- plt.imshow('2차원 픽셀 배열')

```python
from sklearn.datasets import load_digits

digits = load_digits()     # 0부터 9까지 숫자 이미지 데이터
x = digits.images[0]
x
```

```python
plt.imshow(x)
plt.show()
```

```python
# 여러개 이미지 출력

fig, axes = plt.subplots(2, 5, figsize = (12, 6))

for i in range(10):
    axes[i // 5][i % 5].imshow(digits.images[i])
plt.show()
```

## surface(3D 그래프)

- axes.plot_surface(x, y, z)

```python
# 3차원 데이터 생성

from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(50, 150, 24)
y = np.linspace(50, 150, 24)

x, y = np.meshgrid(x, y)     # x와 y의 shape을 합침
z = (x - 100)**2 / y     # z값(2차원 배열)

fig = plt.figure(figsize = (10, 5))
axes = Axes3D(fig)     # 3차원 axes 생성
axes.plot_surface(x, y, z, cmap = 'coolwarm')
plt.title('3D 그래프 시각화')
plt.show()
```
