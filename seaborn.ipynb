```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl
import numpy as np
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
from matplotlib import rcParams
```

```python
sns.set(style = 'darkgrid')     # 배경을 darkgrid 설정

titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
```

```python
# countplot: 항목의 개수

plt.figure(figsize = (10, 6))
sns.countplot(x = 'class', hue = 'who', data = titanic)
plt.show()
```

```python
# lmplot: 선형 회귀 관계

plt.figure(figsize = (10, 6))
sns.lmplot(x = 'total_bill', y = 'tip', height = 8, data = tips)
plt.show()
```

```python
# jointplot: 산점도 + 히스토그램

plt.figure(figsize = (10, 6))
sns.jointplot(x = 'total_bill', y = 'tip', height = 8, data = tips)
plt.show()
```

```python
# heatmap

plt.figure(figsize = (10, 6))
sns.heatmap(titanic.corr(), annot = True)
plt.show()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
# 왼쪽 matplotlib, 오른쪽 seaborn

fig, axes = plt.subplots(ncols = 2, figsize = (10, 5))

# axes[0]: matplotlib
islands = penguins['island'].unique()
color = 0
for island in islands:
    axes[0].scatter(penguins['bill_length_mm'].loc[penguins['island'] == island],
                   penguins['bill_depth_mm'].loc[penguins['island'] == island],
                   c = f'C{color}', label = island)
    color = color + 1
    
axes[0].set_xlabel('bill_length_mm')
axes[0].set_ylabel('bill_depth_mm')

# axes[1]: seaborn
sns.scatterplot('bill_length_mm', 'bill_depth_mm', hue = 'island', data = penguins, ax = axes[1])

plt.show()
```

```python
# 왼쪽: matplotlib 산점도 + seaborn 추세선 / 오른쪽: seaborn 산점도 + matplotlib 중심점

fig, axes = plt.subplots(ncols = 2, figsize = (10, 5))

# axes[0]: matplotlib 산점도
islands = penguins['island'].unique()
color = 0
for island in islands:
    axes[0].scatter(penguins['bill_length_mm'].loc[penguins['island'] == island],
                   penguins['bill_depth_mm'].loc[penguins['island'] == island],
                   c = f'C{color}', label = island)
    color = color + 1
    
    # seaborn 추세선
    sns.regplot('bill_length_mm', 'bill_depth_mm', data = penguins.loc[penguins['island'] == island],
               scatter = False, ax = axes[0])
    
axes[0].set_xlabel('bill_length_mm')
axes[0].set_ylabel('bill_depth_mm')

# axes[1]: seaborn 산점도
sns.scatterplot('bill_length_mm', 'bill_depth_mm', hue = 'island', data = penguins, ax = axes[1])

# matplotlib 중심점
for island in islands:
    axes[1].scatter(penguins['bill_length_mm'].loc[penguins['island'] == island].mean(),
                   penguins['bill_depth_mm'].loc[penguins['island'] == island].mean(),
                   c = 'black', marker = '*', s = 500)
plt.show()
```

## relplot()

- lineplot + scatterplot
- hue 옵션을 통해 색상 구분 / col 옵션을 통해 다중 그래프

```python
sns.set(style = 'darkgrid')     # 배경 설정
plt.rc('font', size = 15)     # 폰트 사이즈 변경
```

```python
fmri = sns.load_dataset('fmri')
tips = sns.load_dataset('tips')
```

```python
# 라인 그래프(신뢰구간 o)

sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri)
```

```python
# 라인 그래프(신뢰구간 x)

sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci = None)
```

```python
# 산점도

sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips)
```

```python
# 산점도(컬럼 나누기)

sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'time', data = tips)
```

## lmplot()

- linear model plot
- 데이터에 적합된 회귀선 표현

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# 요일 구분

sns.lmplot(x = 'total_bill', y = 'tip', hue = 'day', col = 'day', data = tips)
```

```python
# 다른 추세선

sns.lmplot(x = 'total_bill', y = 'tip', lowess = True, data = tips)
```

## distplot()

## catplot()

- categorical plot
- 카테고리형 변수를 시각화할 때 사용

## pairplot()

- 데이터 차원이 여러개인 경우, 한번에 산점도와 히스토그램 확인
- 데이터의 상관관계 확인

```python
penguins = sns.load_dataset('penguins')
sns.pairplot(penguins)
```

```python
sns.pairplot(penguins, hue = 'species')
```

```python
sns.pairplot(penguins, hue = 'species', corner = True)
```

## jointplot()

- 산점도 + 히스토그램

```python
sns.jointplot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm')
```

```python
sns.jointplot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species')
```

```python
sns.jointplot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species', kind = 'kde')
```
